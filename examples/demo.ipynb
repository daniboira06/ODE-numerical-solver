# ODE Numerical Solver - Demo
Notebook de demostración de los cuatro sistemas físicos implementados.

## Imports

```
import sys
sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
from odes import (oscilador_armonico, oscilador_armonico_analitica,
                  pendulo, pendulo_analitica,
                  lorenz, lotka_volterra)
from solvers import euler, rk4, rk45
from plots import (plot_solucion, plot_error, plot_fase,
                   plot_lorenz, plot_lotka_volterra, animate_oscilador)
```

## 1. Oscilador armónico

```
t0, dt, N = 0.0, 0.01, 1000
y0 = np.array([1.0, 0.0])
f  = lambda t, y: oscilador_armonico(t, y, omega=1.0)

t_euler, y_euler = euler(f, t0, y0, dt, N)
t_rk4,   y_rk4   = rk4(f,   t0, y0, dt, N)
t_rk45,  y_rk45  = rk45(f,  t0, y0, t_end=N*dt)

y_ana = oscilador_armonico_analitica(t_rk4, 1.0, 0.0, omega=1.0)
plot_solucion(t_rk4, y_rk4, y_ana, labels=['x', 'v'], title='Oscilador - RK4')
plot_error(t_euler, y_euler, oscilador_armonico_analitica(t_euler, 1.0, 0.0), title='Error - Euler')
plot_error(t_rk4,   y_rk4,   y_ana, title='Error - RK4')
plt.show()
```

## 2. Péndulo simple

```
theta0 = np.radians(10)
y0 = np.array([theta0, 0.0])
f  = lambda t, y: pendulo(t, y, g=9.8, L=1.0)

t_arr, y_arr = rk4(f, t0, y0, dt, N)
y_ana = pendulo_analitica(t_arr, theta0, g=9.8, L=1.0)

plot_solucion(t_arr, y_arr, y_ana, labels=['θ', 'ω'], title='Péndulo - RK4')
plot_error(t_arr, y_arr, y_ana, title='Error péndulo - RK4')
plot_fase(y_arr, labels=('θ', 'ω'), title='Espacio de fases - Péndulo')
plt.show()
```

## 3. Sistema de Lorenz

```
y0 = np.array([1.0, 1.0, 1.0])
f  = lambda t, y: lorenz(t, y, sigma=10.0, rho=28.0, beta=8/3)

t_arr, y_arr = rk4(f, t0, y0, dt, N=5000)

plot_lorenz(y_arr, title='Atractor de Lorenz - RK4')
plot_solucion(t_arr, y_arr, labels=['x', 'y', 'z'], title='Lorenz - RK4')
plt.show()
```

## 4. Lotka-Volterra

```
y0 = np.array([40.0, 9.0])
f  = lambda t, y: lotka_volterra(t, y, alpha=1.0, beta=0.1, delta=0.075, gamma=1.5)

t_arr, y_arr = rk4(f, t0, y0, dt, N=3000)

plot_lotka_volterra(t_arr, y_arr, title='Lotka-Volterra - RK4')
plt.show()
```
